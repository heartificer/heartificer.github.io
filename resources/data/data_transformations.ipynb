```python
import pandas as pd
```

```python
raw = pd.read_csv('usretechnicalpotential.csv')
```

```python
raw.head()
```

```python
raw.columns
```

```python
raw = raw.rename(columns={'Unnamed: 0': 'Region'})
raw
```

```python
types = ['PV', 'Wind', 'CSP', 'biopower', 'Hydrothermal', 'Geothermal', 'hydropower']
for t in types:
    related_cols = [c for c in raw.columns if ((t in c) & c.endswith('GW') & ('rooftop' not in c))]
    print(related_cols)
    raw['{}'.format(t.lower())] = raw[related_cols].sum(axis=1)
```

```python
raw.head()
```

```python
agg_natl = raw.sum()
agg_natl['Region'] = 'National'
agg_natl
```

```python
raw.loc[len(raw.index)] = agg_natl
raw
```

```python
raw.to_csv('usretechnicalpotential_column_aggs.csv')
```

```python
raw.loc[raw.Region== "National", [c for c in raw.columns if c.startswith("all")]].to_csv('usretechnicalpotential_national_column_aggs.csv')
```
